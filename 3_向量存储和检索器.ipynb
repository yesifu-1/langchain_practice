```python
from ollama import Client  #配置ollama客户端
import openai

# 配置 Ollama 客户端
ollama_client = Client("nomic-embed-text")
```

```python
texts = [
    "Dogs are loyal and friendly.",
    "Cats are independent and love their space.",
]

# 获取嵌入向量
embeddings = []
for text in texts:
    response = ollama.embeddings(model='nomic-embed-text', prompt=text)
    embeddings.append(response['embedding'])
    
```
